```python
from transformers import MBartTokenizer, MBartForConditionalGeneration

model_name = "facebook/mbart-large-50-one-to-many-mmt"
tokenizer = MBartTokenizer.from_pretrained(model_name)
model = MBartForConditionalGeneration.from_pretrained(model_name)
```

```python
tokenizer(['My Name is Vedant','I am happy.'])
```

```python
def lang_trans(src='en_XX',out_lang='hi_IN',text=None):
  tokenizer.src_lang=src
  encoded_en=tokenizer(text,return_tensors='pt')

  generated_tokens=model.generate(
    **encoded_en,
    forced_bos_token_id=tokenizer.lang_code_to_id['hi_IN']
    )

  return tokenizer.batch_decode(generated_tokens,skip_special_tokens=True)
```

```python
lang_trans(text='This is very good')
```

```python
text='''The swearing-in ceremony at Raj Bhavan witnessed the initiation of a revamped team, with key players including Samrat Choudhary'''
```

```python
lang_trans(src='en_XX',out_lang='hi_IN',text=text)
```
